## 5 - Descobrir quais so os bairros que no tem nome de Rua. (Vivi)

```python
# Descobrindo nome dos bairros que possuem observaes sem o nome de Rua no DF.
sem_rua = dados['Rua'].isnull()  


dados_sem_rua = dados[sem_rua]
dados_sem_rua['Bairro'].unique()
```

```python
# Percentual dos bairros que possuem alguma observao sem nome de Rua no DF.
print(f"Quantidade de observaes sem informao da rua: {sum(sem_rua)}")

num_bairros_sem_rua = len(dados_sem_rua['Bairro'].unique())
total_bairros = len(dados['Bairro'].unique())
percentual_bairros_sem_rua = round((num_bairros_sem_rua / total_bairros)*100)
print(f"O percentual de bairros que tem uma rua sem nome no banco de dados  de {percentual_bairros_sem_rua}%")
```

Conceitos ensinados:
- Separar informao de uma coluna em formato string;
- Converter string em float;
- Fazer grfico com a biblioteca seaborn;

```python
#Ver a informao do valor na primeira observao, lembrar que python comea do 0.
dados["Valor"][0]
```

```python
# Funo split, usada para separar strings
dados["Valor"][0].split()
```

```python
# Fazer o split para toda a coluna
dados["Valor"].str.split()
```

```python
#Fazer com que o split divida em colunas
dados["Valor"].str.split(expand=True)
```

```python
# Ver os valores nicos
dados["Valor"].str.split(expand=True)[0].unique()
```

```python
# Quantidade de valores diferentes na coluna
len(dados["Valor"].str.split(expand=True)[1].unique())
```

```python
# Quantidade de informaes diferentes na coluna 3 do split de valores
dados["Valor"].str.split(expand=True)[2].unique()
```

```python
# Adicionando no banco de dados (dados) as trs colunas criadas no split
dados[["Moeda","Valor_anuncio", "Tipo_anuncio"]] = dados["Valor"].str.split(expand=True)
dados.head()
```

```python
#Retirar as observaes que estavam em formato de aluguel
dados_vendas = dados[dados["Tipo_anuncio"].isnull()]
dados_vendas
```

```python
dados_vendas.info()
```

```python
#Como os valores esto usando ponto no formato brasileiro  preciso corrigi lo para deixar a variavel tipo numerica(inteiro ou float)

dados_vendas["Valor_anuncio"].str.replace(".","").astype(float)
```

```python
dados_vendas["Valor_anuncio_float"] = dados_vendas["Valor_anuncio"].str.replace(".","").astype(float)
```

```python
dados_vendas.info()
```

```python
#plotar histograma de valor de anuncio
dados_vendas['Valor_anuncio_float'].plot.hist()
```

```python
#plotar histograma de valor de anuncio
dados_vendas['Valor_anuncio_float'].plot.hist(bins=50)
```

```python
dados_vendas.describe()
```

```python
# Mudar a quantidade de casas decimais que aparecem 
pd.set_option("display.precision",2)
dados_vendas.describe()
```

```python
# Outra maneira de alterar a quantidade de casas decimais que aparecem 
pd.set_option("display.float_format", lambda x: '%.2f' % x)
dados_vendas.describe()
```

```python
#Incluir mais estatisticas no describe
dados_vendas.describe(include="all")
```

```python
#Describe quando as variaveis so do tipo objeto/string
dados_vendas[["Rua","Bairro","Cidade"]].describe()
```
